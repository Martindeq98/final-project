## Setting
AR(1) model

$$X_{t} = a X_{t - 1} + \varepsilon_t, \qquad X_1 \sim \mathcal{N}\left(0, \frac{1}{1 - a^2}\right).$$

$$H_0: a_\text{true} = a, \qquad H_1: a_\text{true} \neq a.$$

We use cross-validation to distinguish between the two.

\begin{align*}
CV(a = a_\text{true}) 
&= \frac{1}{T - 1}\sum_{t = 1}^{T - 1}\left(X_{t+1} - a_\text{true} X_t\right)^2 \\
&= \frac{1}{T - 1}\sum_{t = 1}^{T - 1}\left(a_\text{true} X_{t} + \varepsilon_{t+1} - a_\text{true} X_t\right)^2 \\
&= \frac{1}{T - 1}\sum_{t = 1}^{T - 1}\left(\varepsilon_{t+1}\right)^2 \sim \frac{\chi^2_{T-1}}{T - 1}\\
\end{align*}

\begin{align*}
CV(a \neq a_\text{true}) 
&= \frac{1}{T - 1}\sum_{t = 1}^{T - 1}\left(X_{t+1} - \hat{a}^{-t} X_t\right)^2 \\
&= \frac{1}{T - 1}\sum_{t = 1}^{T - 1}\left(a_\text{true} X_{t} + \varepsilon_{t+1} - \hat{a}^{-t} X_t\right)^2 \\
&= \frac{1}{T - 1}\sum_{t = 1}^{T - 1}\left(\varepsilon_{t+1} + \left(a_\text{true} - \hat{a}^{-t}\right)X_t\right)^2\\
\end{align*}

In the limit, we have that
\begin{align*}
CV(0) \sim \frac{\chi^2_{T-1}}{T - 1} &\overset{T \rightarrow \infty}{\rightarrow} \mathcal{N}\left(1, \frac{2}{T - 1}\right)\\
CV(1) &\overset{T \rightarrow \infty}{\rightarrow} \mathcal{N}\left(1, \frac{2}{T - 1}\right)\\
\end{align*}
