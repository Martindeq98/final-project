```python
from gsq.gsq_testdata import bin_data, dis_data
import numpy as np
from statsmodels.tsa.api import VAR
import itertools
import seaborn as sns
import matplotlib.pyplot as plt

# read dataset with the variables
NUM_VAR = 5
LENGTH = 5000

# Acquire binary data
data = np.array(bin_data).reshape((LENGTH, NUM_VAR))
```

```python
def plot_heatmap(data, signif = 0.1, NUM_VAR = len(data[0])):
    
    # Create model
    model = VAR(data)

    # Fit model order 1
    result = model.fit(1)

    # Create heapmap
    heatmap1 = np.array(np.zeros((NUM_VAR, NUM_VAR)))

    for a, b in itertools.product(range(NUM_VAR), repeat=2):   
        heatmap1[b][a] = result.test_causality(a, b, signif=signif, kind = "f").pvalue

    # Plot heatmap based on conclusion
    heatmap_plot = sns.heatmap(heatmap1, xticklabels="12345", yticklabels="12345", annot=True)
    heatmap_plot.set(xlabel="Target (causee)", ylabel="Source (causer)")
    plt.title("P-values of granger causality using F test")
    plt.show()
                 
    return heatmap1

heatmap = plot_heatmap(data)
```

```python
import networkx as nx

def draw_network(heatmap, signif = 0.4):
    # Draw graph
    rows, cols = np.where(heatmap < signif)
    edges = zip(rows.tolist(), cols.tolist())

    gr = nx.DiGraph()
    gr.add_nodes_from(range(NUM_VAR))
    gr.add_edges_from(edges, arrowstyle="->", arrowsize=10,)
    plt.title("Inferred causal graph")
    
    nx.draw(gr, with_labels=True, labels=dict(zip(range(NUM_VAR), dict(zip(range(NUM_VAR), ["1", "2", "3", "4", "5"])))))

draw_network(heatmap)
```

```python
NUM_VAR = 5
LENGTH = 10000

data = np.array(dis_data).reshape((LENGTH, NUM_VAR))
```

```python
# Import Statsmodels
from statsmodels.tsa.api import VAR
import itertools
import seaborn as sns
import matplotlib.pyplot as plt

# Create model
model = VAR(data)

# Fit model order 1
result = model.fit(1)

# Significance
signif = 0.21

# Create heapmap
heatmap1 = np.array(np.zeros((NUM_VAR, NUM_VAR)))

for a, b in itertools.product(range(NUM_VAR), repeat=2):   
    heatmap1[b][a] = result.test_causality(a, b, signif=signif, kind = "f").pvalue

# Plot heatmap based on conclusion
heatmap_plot = sns.heatmap(heatmap1, xticklabels="12345", yticklabels="12345", annot=True)
heatmap_plot.set(xlabel="Target (causee)", ylabel="Source (causer)")
plt.title("P-values of granger causality using F test")
plt.show()
```

```python
import networkx as nx

# Draw graph
rows, cols = np.where(heatmap1 < signif)
edges = zip(rows.tolist(), cols.tolist())


gr = nx.DiGraph()
gr.add_nodes_from(range(NUM_VAR))
gr.add_edges_from(edges, arrowstyle="->", arrowsize=.001,)
plt.title("Inferred causal graph")
nx.draw(gr, with_labels=True, labels=dict(zip(range(NUM_VAR), dict(zip(range(NUM_VAR), ["1", "2", "3", "4", "5"])))))
```

```python
import scipy

nEval = 0

def reg_minimizer(data, l = 0.001):

    # Other approach: Minimize the two-norm plus some regularizer
    def f(A, l = l):
        B = A.reshape((NUM_VAR, NUM_VAR))
        cost = 0
        for t in range(1, 1000):
            cost += np.linalg.norm((data[t] - np.matmul(B, data[t-1]))) ** 2
        return cost + l * sum(np.absolute(A))

    def callbackF(Xi):
        global nEval
        print('eval_number: {0:4d}, ..., '.format(nEval), end="")
        nEval += 1
    
    return scipy.optimize.minimize(f, np.ones(NUM_VAR * NUM_VAR), callback=callbackF)

results = reg_minimizer(data)
```

```python
print(results.x)
B_est = np.round(results.x, 1).reshape((NUM_VAR,NUM_VAR)).transpose()

# Draw graph
rows, cols = np.where(abs(np.array(B_est)) > 0.2)
edges = zip(rows.tolist(), cols.tolist())

gr = nx.DiGraph()
gr.add_nodes_from(range(NUM_VAR))
gr.add_edges_from(edges, arrowstyle="->", arrowsize=.001,)
plt.title("Inferred causal graph")
nx.draw(gr, with_labels=True, labels=dict(zip(range(NUM_VAR), dict(zip(range(NUM_VAR), ["1", "2", "3", "4", "5"])))))
```
