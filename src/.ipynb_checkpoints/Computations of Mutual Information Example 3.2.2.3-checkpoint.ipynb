# Computations of Mutual Information Example 3.2.2.3


```python
from math import log2

# Computing H_type
H_type = -9 / 17 * log2(9 / 17) - 8 / 17 *log2(8 / 17)

# Computing H_colour
H_colour = -8 / 17 * log2(8 / 17) - 6 / 17 * log2(6 / 17) - 3 / 17 * log2(3 / 17)

# Computing H_stone
H_stone = -5 / 17 * log2(5 / 17) - 12 / 17 * log2(12 / 17)

# Compute H_joint
H_joint =  -2 / 17 * log2(2 / 17) - 2 / 17 * log2(2 / 17) - 3 / 17 * log2(3 / 17) 
H_joint += - 2 / 17 * log2(2 / 17) - 1 / 17 * log2(1 / 17) - 3 / 17 * log2(3 / 17) 
H_joint += - 3 / 17 * log2(3 / 17) - 1 / 17 * log2(1 / 17)

# multi information
multiMI = H_type + H_stone + H_colour - H_joint

# print results
print(f"H_type: {H_type:.3f}, H_colour: {H_colour:.3f}, H_stone: {H_stone:.3f}, H_joint: {H_joint:.3f}, multi-MI: {multiMI:.3f}")
```

```python
# joint entropy of type and stone
H_typestone = - 4 / 17 * log2(4 / 17) - 5 / 17 * log2(5 / 17) - 1/ 17 * log2(1 / 17) - 7 / 17 * log2(7 / 17)

# mutual information of type and stone
I_typestone = H_type + H_stone - H_typestone

# information component of type = fruit and stone = yes
i_fruitstone = log2((4 / 5) / (9 / 17))

# information component of type = veg and stone = yes
i_vegstone = log2((1 / 5) / (8 / 17))

print(f"H_typestone: {H_typestone:.3f}, I_typestone: {I_typestone:.3f}, i_fruitstone: {i_fruitstone:.3f}, i_vegstone: {i_vegstone:.3f}")
```
