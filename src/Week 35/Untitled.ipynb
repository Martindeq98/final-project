```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib

matplotlib.use("pgf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
    'font.size': 8
})

matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
n = 2500
#np.random.seed(seed = 1)
plt.figure(figsize = (3.0, 2.5))
plt.ylabel(r"$Y$")
plt.xlabel(r"$X$")
plt.title(r"Model with Gaussian noise where $Y = 0.8X$")
X = np.random.normal(0, 1, n)
Y = 0.8 * X + np.random.normal(0, 0.6, n)
#help(plt.scatter)
plt.scatter(X, Y, marker = "o", s = 8)
plt.savefig("scatter_gauss_1.pdf", bbox_inches = "tight")
print(np.round((np.mean(X), np.cov(X), np.mean(Y), np.cov(Y)), 3))
```

```python
Z = np.random.multivariate_normal(np.zeros(2), np.cov(X, Y), 100000)
plt.scatter(Z.T[0], Z.T[1], s = 10)
```

```python
# np.random.seed(seed = 1)
plt.figure(figsize = (3.0, 2.5))

plt.ylabel(r"$Y$")
plt.xlabel(r"$X$")
plt.title(r"Model with Gaussian noise where $X = 0.8Y$")

Y = np.random.normal(0, 1, n)
X = 0.8 * Y + np.random.normal(0, 0.6, n)

plt.scatter(X, Y, s = 8)
plt.savefig("scatter_gauss_2.pdf", bbox_inches = "tight")
print(np.round((np.mean(X), np.cov(X), np.mean(Y), np.cov(Y)), 3))
```

```python
# n = 2500
plt.figure(figsize = (3.0, 2.5))

plt.ylabel(r"$Y$")
plt.xlabel(r"$X$")
plt.title(r"Model with uniform noise where $Y = 0.8X$")

plt.xlim(-2, 2)
plt.ylim(-3, 3)
a = np.sqrt(3)
b = 0.6 * np.sqrt(3)
X = np.random.uniform(-a, a, n)
Y = 0.8 * X + np.random.uniform(-b, b, n)
plt.scatter(X, Y, s = 8)
print(np.round((np.mean(X), np.cov(X), np.mean(Y), np.cov(Y)), 3))

# plt.savefig("scatter_unif_1.pgf")
# x = np.linspace(min(X), max(X), 100)
# plt.plot(x, x * 0.8, color = 'black', label = "Y = 0.8X")

# x = np.linspace(min(X), max(X), 100)
# plt.plot(x, x / 0.8, color = 'red', label = "X = 0.8Y")

# plt.legend()
```

```python
n = 1000

a = np.sqrt(3)
b = 0.6 * np.sqrt(3)
X = np.random.uniform(-a, a, n)
Y = 0.8 * X + np.random.uniform(-b, b, n)

print(1 / n * np.linalg.norm(X - (Y - 0.8 * X)) ** 2)
print(1 / n * np.linalg.norm(Y - (X - 0.8 * Y)) ** 2)
```

```python
n = 1000
plt.figure(figsize = (2.0, 2.0)) 

plt.ylabel(r"$Y$")
plt.xlabel(r"$X$")
plt.title(r"Uniform noise where $Y = 0.8X$")

plt.scatter(X, Y, s = 8)

x = np.linspace(min(X), max(X), 100)
plt.plot(x, x / 0.8, color = 'red', label = "X = 0.8Y", linewidth = 3)

x = np.linspace(min(X), max(X), 100)
plt.plot(x, x * 0.8, color = 'black', label = "Y = 0.8X", linewidth = 3)

plt.legend()
plt.savefig("scatter_unif_1_lines1.pgf", bbox_inches = "tight")
plt.show()

plt.figure(figsize = (2.0, 2.0))
plt.xlabel(r"$X$")
plt.ylabel(r"Residual: $Y - 0.8X$")
plt.title(r"Residuals when $X \rightarrow Y$")
plt.scatter(X, Y - 0.8 * X, s = 8)
plt.savefig("scatter_unif_resid_11.pgf", bbox_inches = "tight")
plt.show()

plt.figure(figsize = (2.0, 2.0))
plt.xlabel(r"Residual: $X - 0.8Y$")
plt.ylabel(r"$Y$")
plt.title(r"Residuals when $Y \rightarrow X$")
plt.scatter(X - 0.8 * Y, Y, s = 8)
plt.savefig("scatter_unif_resid_21.pgf", bbox_inches = "tight")
```
