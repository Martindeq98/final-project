
# Generate Data

```python
import networkx as nx
import random
import numpy as np

def generate_data(NODES, n, edge_prob = 0.15, verbose = True):
    S0 = 2 * NODES
    G = nx.gnp_random_graph(NODES, edge_prob, directed=True)
    DAG = nx.DiGraph([(u,v,{'weight': random.choice([-1, 1]) * random.uniform(1, 5)}) for (u,v) in G.edges() if u < v])

    while len(DAG.edges) > S0 :
        edge_to_remove = random.choice(list(DAG.edges()))
        DAG.remove_edge(edge_to_remove[0], edge_to_remove[1])

    
    # nx.draw(DAG, labels=dict(zip(range(NODES), range(NODES))))

    # Generate weighted matrix B
    B = nx.adjacency_matrix(DAG).todense()


    # Create permutation
    top_list = list(nx.topological_sort(DAG))

    P = np.zeros((NODES, NODES))
    for i in range(len(B)):
        for j in range(len(B)):
            if top_list[i] == j:
                P[i][j] = 1

    # Generate omegas
    omegas = np.diag(np.random.uniform(0.1, 0.2, NODES))
        
    # Generate sigma
    sigma = 0.7 * np.ones((n, n)) + 0.3 * np.identity((n))
    
    # compute psi
    matmul1 = np.linalg.inv(np.identity(NODES) - B).transpose()
    matmul2 = omegas
    matmul3 = np.linalg.inv(np.identity(NODES) - B)

    psi = np.matmul(np.matmul(matmul1, matmul2), matmul3)

    # compute cov
    cov = np.kron(sigma, psi) # np.matmul(P.transpose(), np.matmul(psi, P)))

    # generate data
    X = np.random.multivariate_normal(mean=np.zeros(NODES * n), cov=cov).reshape((n, NODES))
    
    if verbose:
        print(f"Generating DAG with {NODES} nodes and {S0} edges.")
        print(f"Weighted matrix B*:\n{np.round(B,1)}.")
        print(f"Computed omegas:\n{np.round(omegas, 1)}.")
        print(f"Computed sigma:\n{sigma}.")
        
    return X, B, omegas, sigma

NODES, n = 50, 200
X, B, omegas, sigma = generate_data(NODES, n, verbose = False)
```
