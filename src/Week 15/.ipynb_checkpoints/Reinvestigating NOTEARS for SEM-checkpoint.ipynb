### NOTEARS on VAR, with diagonals

```python
reload(linear)

# set random seed
utils.set_random_seed(1)

n, d, s0, graph_type, sem_type = 100, 5, 5, 'ER', 'gauss'
B_true = utils.simulate_dag(d, s0, graph_type)

B_true[0][-1] = 1
B_true[-1][0] = 1

W_true = utils.simulate_parameter(B_true, w_ranges = ((-0.5, -0.25), (0.25, 0.50)))

for i in range(d):
    B_true[i][i] = 1
    W_true[i][i] = B_true[i][i] * np.random.uniform(0.3, 0.7)
    
X = utils.simulate_var(W_true, n)
W_notears = linear.notears_linear(X, lambda1=0.0, loss_type='l2-var', w_threshold = 0.01, h_tol = 1e-16)

# plot X
plt.plot(X);

# print W
print(f"True W:\n{np.round(W_true, 2)}.\n")
print(f"NOTEARS W:\n{np.round(W_est, 2)}.\n")

acc = utils.count_accuracy(B_true, W_est != 0)
print(acc)
```
