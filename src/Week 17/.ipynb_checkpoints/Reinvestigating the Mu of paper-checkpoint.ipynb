### What should $\mu$ be?
In general, $\mu$ refers to estimating the coefficients that should be zero.
$$\mu_{X}(\bar{F})=\max _{j \notin \bar{F}}\left\|\left(X_{F}^{T} X_{\bar{F}}\right)^{-1} X_{\bar{F}}^{T} \mathbf{x}_{j}\right\|_{1}.$$
In other words, "The maximization occurs over atoms that do not participate in the optimal representation of the signal."

We have the optimal representation $X_{\bar{F}}$ of our signal, and we regress our non-optimal atoms on it. The non-optimal atoms are where $W = 0$, as they do not contribute to the signal.

For example, if $W_{i, j} = 0$, then the signal $X_i$ does not contribute to the signal $X_j$. We check what happens when we regress $X_j$ on the optimal signal generation for $X_i$. The optimal signal generation is the signals $X_k$ that are non-zero in the $i$th column of $W$.

```python
X_pred = X_norm
X_val = Y_norm
print(np.shape(X_pred))
print(W_true)

for i in range(np.shape(W_true)[0]):
    col = W_true[:, i]
    
    # get non-zero coefficients
    #print(col.nonzero())  
    
    # get only the atoms that participate in the optimal representation
    X_F_bar = X_pred[:, col.nonzero()[0]]
    #print(np.shape(X_F_bar))
    
    # regress all zero coefficients against the optimal representation
    zero_coefs = np.where(col == 0)[0]
    #print(zero_coefs)
    
    for index in zero_coefs:
        print(np.linalg.inv(X_F_bar.T @ X_F_bar) @ X_F_bar.T @ X_val[:, index])
    
```

### Counter Example showing why $\mu$ is not applicable

$$X_{t, 1} = aX_{t - 1, 1} + \text{noise}, |a| < 1.$$
$$X_{t, 2} = c X_{t - 1, 1} + \text{noise}, c \in \mathbb{R}.\ $$
We have
$$\mu_{X}(\bar{F})=\max_{j \notin \bar{F}}\left\|\left(X_{F}^{T} X_{\bar{F}}\right)^{-1} X_{\bar{F}}^{T} \mathbf{x}_{j}\right\|_{1}.$$
Now, only $X_1$ is in the optimal representation. Hence, $X_{\bar{F}} = X_1$. Now, the indices that are zero are:
- Using $X_2$ to predict $X_1$. According to the $\mu$ formula, we use $X_{\bar{F}} = X_1$ to predict $X_2$. However, we have in our formula that $X_2 = cX_1$ (+ noise). Hence, we expect the ourcome here to be approximately $c$.
- Using $X_2$ to predict $X_2$. According to the $\mu$ formula, we use $X_{\bar{F}} = X_1$ to predict $X_2$. However, we have in our formula that $X_2 = cX_1$ (+ noise). Hence, we expect the ourcome here to be approximately $c$.

Now $c$ can get arbitrarily big, even without causing problems to the VAR(1) model. Hence, this $\mu$ can get arbitrarily big as well! However, the formula required $X$ to be normalized, and when we normalize, the value for $\mu$ tends to be around $1$. However, this means that we cannot give any guarantees using this proof method.

This is strange, as there does not seem to be much wrong with the way to model is defined, OMP should give good results.

```python
a, c = -0.5, 0.3
n, T = 3, 100

W_true = np.array([[a, c, 0], [0.0, 0.0, 0], [0, 0, 0]])
print(f"W:\n{np.round(W_true, 2)}.")

X = h.generate_var_2(T, n, W_true, np.identity(n))
plt.plot(X);
```

```python
def mu_matrix(W, X):
    
    mu = -np.inf
    
    for i in range(np.shape(W_true)[0]):
        # get column
        col = W[:, i]
    
        # get only the atoms that participate in the optimal representation
        X_F_bar = X[:-1, col.nonzero()[0]]
    
        # regress all zero coefficients against the optimal representation
        zero_coefs = np.where(col == 0)[0]
    
        for index in zero_coefs:
            mu = max(mu, np.linalg.inv(X_F_bar.T @ X_F_bar) @ X_F_bar.T @ X[1:, index])
            
    return mu[0]

# Mu on unnormalized X
print(f"Mu on unnormalized X: {round(mu_matrix(W_true, X), 2)}.")

# Normalize
X_norm = X / np.sqrt((X ** 2).sum(axis=0) / T)

# Mu on normalized X
print(f"Mu on unnormalized X: {round(mu_matrix(W_true, X_norm), 2)}.")
```

We see indeed that our expectations were true. $\mu = c$ for the unnormalized quantity, and very close to $1$ for the normalized quantity. When $c < 1$, the normalized quantity will lie between $c$ and $1$, but nowhere as close to one as the noise is now a relatively larger contributor.

### Why does OMP still give good results?
Interestingly, OMP still gives very good results. The thing that goes almost wrong when our $\mu$ is large is deciding which variable to pick. In the beginning, two coefficients both yield almost equivalent gains, whereas only one is correct, and the other is incorrect. Nevertheless, the gain of the correct variable always seems to be just slightly larger. When this variable has been added, adding the other variable does not result in any larger gain anymore.

```python
def update_betas2(F, Y):
    """Compute OLS coefficients using only the entries in F"""
    
    # fit only using coefficients in F
    X_F = np.kron(np.eye(n, dtype=float), X[:-1])[:, F]
    
    # fit
    ols_fit = OLS(Y.T.reshape((T - 1) * n), X_F).fit()
    
    # get parameters
    params = np.zeros(n ** 2)
    params[F] = ols_fit.params
    
    return params.reshape(n, n).T
```

```python
def OMP_matrix_DAG(X, Y, max_coefs = int(1e10), tol = 0.1):
    
    # intialize F
    F = set()
    
    # forbidden indices
    N = set()
    
    # initialize beta
    betas = np.zeros((n, n))
    
    for coef in range(min(n ** 2, max_coefs)):
        
        # get the gains
        gains = np.array([np.abs(normalized(X[:, i]) @ ((X @ betas)[:, j] - Y[:, j])) for j in range(n) for i in range(n)])
        
        # set the forbidden coefficients to negative value, so that they will never be chosen
        gains[list(N)] = -np.ones(len(list(N)))
        
        # find the index that maximizes the gains
        i_max = np.argmax(gains)

        # check if the gain is large enough
        if max(gains) <= tol: break
        
        # add index to F
        F.add(i_max)
        
        # update residuals
        new_betas = update_betas2(list(F), Y)
        print(new_betas)
        
        # check if adding this beta will yield a cycle
        if h.is_dag(new_betas):
            # if not, we add it
            betas = new_betas
        else:
            # if so, we remove it, and add it to the forbidden list
            F.remove(i_max)
            N.add(i_max)
        
        # print for feedback
        print(f"Gain: {round(max(gains), 2)}.")
        print(f"Betas:\n{np.round(betas, 2)}.\n")

    print(N)
    
    # return betas
    return betas.reshape(n, n)

print("W_NORM:\n", W_true, "\n")
T = 1000
X = h.generate_var_2(T, n, W_true, np.identity(n))
W_OMP_paper = OMP_matrix_DAG(X[:-1], X[1:], tol = 0.1)

print("\nW_OMP_PAPER:\n", np.round(W_OMP_paper, 3))
print("\nW_OMP:\n", np.round(m._OMP(X), 3))

h.score(X, m._OMP(X), W_true)
h.score(X, W_OMP_paper, W_true)
```
