### Perform OLS

```python
W_hat = np.zeros((n, n))
# get parameters
for i in range(n):
    W_hat[:, i] = np.linalg.inv(X[:-1].T @ X[:-1]) @ X[:-1].T @ X[1:, i]

print(W_hat)
```

```python
def get_all_violating_edges(W):
    """Returns all the edges of W that violate DAG-ness"""
    W_adj = W.copy()
    W_adj -= np.diag(np.diag(W_adj))
    G = networkx.from_numpy_matrix(W_adj, create_using=networkx.DiGraph)
    
    violators = set()
    for cycle in networkx.simple_cycles(G):
        edges = [(cycle[i], cycle[(i+1)%len(cycle)]) for i in range(len(cycle))]
        for edge in edges:
            violators.add(edge)
                
    return violators
```
