```python
# load stats models
import statsmodels.api as sm
# import granger causality test
from statsmodels.tsa.stattools import grangercausalitytests
# import numpy
import numpy as np
# import matplotlib
import matplotlib.pyplot as plt
# load the data
data = sm.datasets.macrodata.load_pandas()
# get the realgdp and realcons, drop unknown values
data_xy = data.data[["realgdp", "realcons"]].pct_change().dropna()
data_yx = data.data[["realcons", "realgdp"]].pct_change().dropna()
```

```python
plt.plot(data_xy['realcons'], label='realcons')
plt.plot(data_xy['realgdp'], label='realgdp')
plt.legend()
```

```python
gc_res = grangercausalitytests(data_xy, 20)
```

```python
gc_res = grangercausalitytests(data_yx, 20)
```
