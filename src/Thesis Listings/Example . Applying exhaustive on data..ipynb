```python
### Generate the data yourself

## Set random seed
np.random.seed(seed = 12345)

## Number of samples, number of variables
T, n = 100, 3

### Coefficient matrix
A_true = np.array([[0.8, -0.5, 0.5], [0.0, 0.8, 0.0], [0.0, 0.0, 0.8]])

## True permutation
P_true = np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]])

## Data generating W
W_true = P_true.T @ A_true @ P_true
print(f"True W:\n{np.round(W_true, 2)}.\n")

## Generate series
X = h.generate_var_2(T, n, A_true, P_true)
```

```python
## Plot series
fig = plt.figure(figsize = (5.5, 3))
fig.patch.set_alpha(0)
plt.xlabel(r"Timestep $t$")
plt.ylabel(r"Value")
plots = plt.plot(X); 
plt.legend(plots, [r"$X_1$", r"$X_2$", r"$X_3$"], loc = 1);
plt.savefig('figure exhaustive example.pgf')
```
